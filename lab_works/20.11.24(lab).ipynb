```python
import matplotlib.pyplot as plt

# Data for Madhya Pradesh
parties = ['BJP', 'INC', 'BSP', 'OTHERS']
madhyapradesh = [163, 66, 0, 1]
explode_mp = [0.05 if i == max(madhyapradesh) else 0 for i in madhyapradesh]

# Data for Rajasthan
rajasthan = [115, 69, 2, 13]
explode_rj = [0.05 if i == max(rajasthan) else 0 for i in rajasthan]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))

# Pie chart for Madhya Pradesh
ax1.pie(madhyapradesh, explode=explode_mp, labels=parties, autopct='%1.1f%%', shadow=False, startangle=180)
ax1.set_title('Madhya Pradesh')

# Pie chart for Rajasthan
ax2.pie(rajasthan, explode=explode_rj, labels=parties, autopct='%1.1f%%', shadow=False, startangle=180)
ax2.set_title('Rajasthan')

# Display the charts
plt.show()
```

```python

import numpy as np
bar_width = 0.4
index = np.arange(len(parties))

fig, ax = plt.subplots(figsize=(10, 6))

ax.bar(index, madhyapradesh, bar_width, label='Madhya Pradesh')

ax.bar(index + bar_width, rajasthan, bar_width, label='Rajasthan')
ax.set_xlabel('Party')
ax.set_ylabel('Seats Won')
ax.set_title('Results')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(parties)
ax.legend()

# Display the chart
plt.show()
```

```python
def solve(k,n):
    if k >=10:
        val = np.random.randint(0,n+1,size = k)
        return val
    else:
        return "size is less than 10"
k = int(input("inter size"))
n = int(input("enter limit"))
val = solve(k,n)
val
```

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
def is_composite(n):
    if n > 1 and not is_prime(n):
        return True
    return False
lst1 = [int(i) for i in val if is_prime(i)]    
lst2 = [int(i) for i in val if is_composite(i)]   
print(lst1)
print(lst2)
```

```python

lst3 = [i**2 for i in lst1 ]    
lst4 = [float(f"{np.sqrt(i):.2f}") for i in lst2 ] 
print(lst3)
print(lst4)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 7))
ax1.scatter(lst1, lst3, color='blue', label='Prime vs. Square') 
ax1.set_xlabel('Prime Numbers') 
ax1.set_ylabel('Square of Prime Numbers') 
ax1.set_title('Prime Numbers vs. Their Squares') 
ax1.legend() 
ax2.scatter(lst2, lst4, color='red', label='Composite vs. Square Root') 
ax2.set_xlabel('Composite Numbers') 
ax2.set_ylabel('Square Root of Composite Numbers') 
ax2.set_title('Composite Numbers vs. Their Square Roots') 
ax2.legend()
```

```python
import numpy as np
A = np.array([[44, 38, 21],
              [32, 28, 15],
              [24, 20, 10]])
B = np.array([1412, 1024, 728])
solution = np.linalg.solve(A, B)
num_males, num_females, num_kids = solution
print(f"Number of males: {int(num_males)}")
print(f"Number of females: {int(num_females)}")
print(f"Number of kids: {int(num_kids)}")
```

```python
import numpy as np

# Get the shape of the arrays from the user
shape = tuple(map(int, input("Enter the shape of the 3D arrays (e.g., 3 4 5): ").split()))

# Generate two random 3D arrays with the given shape
A1 = np.random.randint(1, 101, size=shape)
A2 = np.random.randint(1, 101, size=shape)

print("Array A1:\n", A1)
print("Array A2:\n", A2)

# Select elements from A1 that are divisible by 5
divisible_by_5 = A1[A1 % 5 == 0]
print("Elements of A1 divisible by 5:\n", divisible_by_5)

# Select elements from A2 that are divisible by 4
divisible_by_4 = A2[A2 % 4 == 0]
print("Elements of A2 divisible by 4:\n", divisible_by_4)

# Join the two arrays to form a single dimensional array
combined_array = np.concatenate((divisible_by_5, divisible_by_4))
print("Combined array:\n", combined_array)

# Find the largest possible square matrix from the combined array
length = len(combined_array)
size = int(np.sqrt(length))
largest_square_matrix = combined_array[:size*size].reshape(size, size)
print("Largest possible square matrix:\n", largest_square_matrix)

```

```python

def Create_Array(*args):
    # Get the number of arguments
    n = len(args)
    
    # Create an n x n matrix filled with zeros
    array = np.zeros((n, n), dtype=int)
    
    # Set the diagonal elements to the values in args
    np.fill_diagonal(array, args)
    
    return array

result = Create_Array(1, 2, 3, 5,4,6)
print(result)
```
